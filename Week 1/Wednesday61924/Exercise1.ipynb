```python
# Using the imdb top 1000 movies dataset find the following 
    # Get rid of useless columns 
    # Oldest movie 
    # Newest movie 
    # Top 10 movies by IMDB rating  
    # Top movie for each genre 
    # Director with the most movies 
    # Star with the most movies  
    # Highest grossing movie for each genre 
    # Lowest grossing movie for each director 
    # Save the dataframe as a parquet file 
    # Aggregate the gross revenue for all 1000 movies. 

import pandas as pd

df = pd.read_csv("imdb_top_1000.csv")
df.head()
```

```python
# Get rid of useless columns
    #cols to remove: poster link, overview, 
df.info()
df.drop("Poster_Link", axis=1, inplace=True)
df.drop("Overview", axis=1, inplace=True)
df.drop("Meta_score", axis=1, inplace=True)

df['Gross'] = df['Gross'].str.replace(',', '').astype(float)
```

```python
# Oldest movie 

#df.sort_values(by="Released_Year", inplace=True)
#df.loc[df["Series_Title"] == "Das Cabinet des Dr. Caligari"]
df.loc[df['Released_Year'].idxmin()]
```

```python
## FOUND A WEIRD ENTRY
# Appollo 13 has a released_year as PG, so I am gong to drop that entry
df.loc[df["Released_Year"] == "PG"]
df.drop([966], inplace=True)
```

```python
# Newest movie 
df.sort_values(by="Released_Year", inplace=True, ascending=False)
#df.head(10) # Shows me there are 6 movies released in 2020
df.head(6)
```

```python
# Top 10 movies by IMDB rating  
df.sort_values(by="IMDB_Rating", ascending=False, inplace=True)
df.head(10)
```

```python
# Top movie for each genre 

# explode on genre
temp1 = df
df.sort_values('Genre', inplace=True)
df["Genre"] = df["Genre"].str.split(', ')
df = df.explode('Genre')
df.dropna()
# group by genre
# sort accending by IMDB rating
#TODO - eliminate entries which are a previous Genre i.e. 2 actions
df.loc[df.groupby('Genre')['IMDB_Rating'].idxmax()]
```

```python
# Director with the most movies 
df["Director"].value_counts().idxmax()
```

```python
# Star with the most movies 
temp = df
df["Stars"] = df[['Star1', 'Star2', 'Star3', 'Star4']].values.tolist()
df.drop(['Star1', 'Star2', 'Star3', 'Star4'], inplace=True, axis=1)
df = df.explode('Stars')
df.dropna()
df["Stars"].value_counts().idxmax()
```
