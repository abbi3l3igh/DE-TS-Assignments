```python
import pandas as pd

# Read in the csv file
spotify = pd.read_csv("songs_normalize.csv")
spotify.head()
```

```python
spotify.nunique()
# Number of Songs in the data set
spotify.info()
```

```python
# What year is the oldest song?
spotify.sort_values(by="year")
    # this shows us the oldest song is from 1998
```

```python
# What is the most popular song in the dataset?
spotify.sort_values(by='popularity', ascending=False)
    # this tells us Sweater Weather is the most popular song with a score of 89
```

```python
# How many unique genres are there?

# Split genres into their own lines
spotify['genre'] = spotify['genre'].str.split(", ")
spotify = spotify.explode('genre')
spotify.dropna()
spotify.head()
# get number of unique genres
spotify['genre'].nunique()
```

```python
# What is the most popular unique individual genre by number of songs on the list?
spotify['genre'].value_counts()
# This tells us that Pop is the most popular
```

```python
# What is the most popular artist by total number of songs on the list?
new_spotify = pd.read_csv("songs_normalize.csv")
new_spotify['artist'].value_counts()
# This shows us Rihana is the most popular Artist by number of songs
```
